# **Step 2: Preprocessing**

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
```

```python
# Combine datasets
combined_data = pd.concat([datasets[f"all_data_{i}"] for i in range(1, 91)], ignore_index=True)

# Filter out noise
filtered_data = combined_data[combined_data["Letter"] != "noise"]

# Define sensor columns
sensor_columns = ['Acc_X', 'Acc_Y', 'Acc_Z', 'Gyro_X', 'Gyro_Y', 'Gyro_Z', 'Mag_X', 'Mag_Y', 'Mag_Z', 'Force1', 'Force2', 'Force3']

# Normalize sensor data (Z-score normalization)
scaler = StandardScaler()
filtered_data[sensor_columns] = scaler.fit_transform(filtered_data[sensor_columns])
```

```python
# Function to group data into sequences by label
def create_sequences(data):
    sequences = []
    labels = []
    current_label = None
    current_sequence = []

    for _, row in data.iterrows():
        if row["Letter"] != current_label:
            # Save the current sequence if it's not empty
            if current_sequence:
                sequences.append(np.array(current_sequence))
                labels.append(current_label)
            # Start a new sequence
            current_label = row["Letter"]
            current_sequence = []

        # Append sensor values to the current sequence
        current_sequence.append(row[sensor_columns].values)

    # Save the final sequence
    if current_sequence:
        sequences.append(np.array(current_sequence))
        labels.append(current_label)

    return np.array(sequences, dtype=object), labels
```

```python
# prompt: shape of sequences and labels

# Create sequences and labels
sequences, labels = create_sequences(filtered_data)

# Print the shapes
print("Sequences shape:", sequences.shape)
print("Labels shape:", len(labels))

for seq, label in zip(sequences, labels):
    print(f"Sequence Shape: {seq.shape}, Label: {label}")
```

```python
# prompt: Based on a given length, i wanna pad (if len < max_length) or resampla sequences (if len > max_length)

def pad_or_resample(sequence, max_length):
    seq_len = len(sequence)
    if seq_len < max_length:
        # Pad the sequence with zeros
        pad_width = max_length - seq_len
        padded_sequence = np.pad(sequence, ((0, pad_width), (0, 0)), mode='constant')
        return padded_sequence
    elif seq_len > max_length:
        # Resample the sequence
        # Use linear interpolation for resampling
        indices = np.linspace(0, seq_len - 1, max_length, dtype=int)
        resampled_sequence = sequence[indices]
        return resampled_sequence
    else:
        return sequence

# Example usage (assuming 'sequences' is defined as in your previous code)
max_length = max_sequence_length  # Set your desired maximum length

processed_sequences = [pad_or_resample(seq, max_length) for seq in sequences]

# Convert the list of processed sequences back to a NumPy array
processed_sequences = np.array(processed_sequences, dtype=np.float32)

# Now 'processed_sequences' contains sequences either padded or resampled to 'max_length'

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(processed_sequences, encoded_labels, test_size=0.2, random_state=42)
```

# **Step 3: Model Design**






```python
from keras.callbacks import EarlyStopping

# Early stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

from tensorflow.keras.callbacks import LearningRateScheduler

# Define a learning rate scheduler function
# def scheduler(epoch, lr):
#   if epoch < 10:
#     return float(lr)
#   else:
#     # return float(lr * tf.math.exp(-0.1))
#     return float(lr * tf.math.exp(-0.05))

# Create the learning rate scheduler callback
# lr_scheduler = LearningRateScheduler(scheduler)

lr_scheduler = tf.keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss', factor=0.3,
    patience=3, min_lr=1e-6
)
```

```python
from tensorflow.keras import layers, models
import tensorflow as tf

# Define Model Input Shape
input_shape = (max_length, len(sensor_columns))  # Time steps x Features
num_classes = len(set(labels))  # Unique number of character classes

# Define CLDNN Model with Softmax Activation
def create_cldnn_cce_model(input_shape, num_classes):
    inputs = layers.Input(shape=input_shape, name='input')

    # Convolutional Layers
    x = layers.Conv1D(512, 5, activation='gelu', padding='same')(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling1D(pool_size=2)(x)
    x = layers.SpatialDropout1D(0.2)(x)

    x = layers.Conv1D(256, 3, activation='gelu', padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling1D(pool_size=2)(x)
    x = layers.SpatialDropout1D(0.2)(x)

    x = layers.Conv1D(128, 3, activation='gelu', padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling1D(pool_size=2)(x)
    x = layers.SpatialDropout1D(0.2)(x)

    # LSTM Layers
    x = layers.Bidirectional(layers.LSTM(64, return_sequences=True, activation='tanh'))(x)
    x = layers.Dropout(0.3)(x)

    x = layers.Bidirectional(layers.LSTM(64, return_sequences=False, activation='tanh'))(x)  # No return_sequences for classification
    x = layers.Dropout(0.3)(x)

    # Fully Connected Layer
    x = layers.Dense(100, activation='gelu')(x)
    x = layers.Dropout(0.3)(x)

    # **Softmax Output Layer for Categorical Cross-Entropy**
    outputs = layers.Dense(num_classes, activation='softmax')(x)

    # Define the model
    model = models.Model(inputs, outputs, name='CLDNN_CCE_Model')
    return model

# Create the CLDNN model
cldnn_cce_model = create_cldnn_cce_model(input_shape, num_classes)

# Save Model
cldnn_cce_model.save('cldnn_cce_model.keras')

# **Compile with Categorical Cross-Entropy Loss**
optimizer = tf.keras.optimizers.AdamW(weight_decay=1e-5, learning_rate=0.001, clipnorm=1.0)
cldnn_cce_model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Display Model Summary
cldnn_cce_model.summary()

# **Train the Model**
cldnn_cce_history = cldnn_cce_model.fit(
    X_train, y_train, validation_data=(X_test, y_test),
    epochs=500,
    callbacks=[early_stopping, lr_scheduler]
)

```

# **Step 4: Prediction**

```python
# Function to Predict a Letter from a Sequence
def predict_letter(sequence, model):
    sequence = np.pad(sequence, ((0, max_length - len(sequence)), (0, 0)), mode='constant')
    sequence = np.expand_dims(sequence, axis=0)
    prediction = model.predict(sequence)
    predicted_label = label_encoder.inverse_transform([np.argmax(prediction)])
    return predicted_label[0]

model = cldnn_cce_model

# Example Prediction
for sample_index in range(100):
  sample = X_test[sample_index]
  print(f"Sample is {sample}")
  predicted_letter = predict_letter(X_test[sample_index], model)
  actual_letter = label_encoder.inverse_transform([y_test[sample_index]])

  print(f"Actual: {actual_letter[0]}, Predicted: {predicted_letter}")
```
